```python
 cylinder = set(d['cyl'] for d in mpg)
```

```python
cylinder
```

```python
ctympgbycyl = []

for c in cylinder:
    sumpg = 0
    cyltypecount = 0
    for d in mpg:
        if d['cyl'] == c:
            sumpg+= float(d['cty'])
            cyltypecount+= 1
    ctympgbycyl.append((c,sumpg/cyltypecount))
ctympgbycyl.sort(key = lambda x:x[0])
    
ctympgbycyl
```

```python
vehicleclass = set(d['class'] for d in mpg)
```

```python
vehicleclass
```

```python
hwybyclass =[]

for t in vehicleclass:
    summpg = 0
    vclasscount = 0
    for d in mpg:
        if d['class']== t:
            summpg+= float(d['hwy'])
            vclasscount+=1
    hwybyclass.append((t,summpg/vclasscount))
hwybyclass.sort(key = lambda x:x[0])

hwybyclass
```
